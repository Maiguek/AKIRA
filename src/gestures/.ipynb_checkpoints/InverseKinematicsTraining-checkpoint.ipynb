```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import matplotlib.pyplot as plt
import mediapipe as mp
from PIL import Image
import pandas as pd
import numpy as np
import random
import serial
import json
import time
import cv2
```

#### Load Servo Data

```python
def load_servo_info(left_servos_file, right_servos_file):
    df_left = pd.read_csv(left_servos_file)
    df_right = pd.read_csv(right_servos_file)
    
    left_servos_list = df_left.to_dict(orient="records")
    right_servos_list = df_right.to_dict(orient="records")
    
    for index in range(len(left_servos_list)):
        left_servos_list[index]["index"] = index
        left_servos_list[index]["current"] = left_servos_list[index]["rest_pos"]
    left_servos_list = {left_servos_list[index]["name"]: left_servos_list[index] for index in range(len(left_servos_list))}
    for index in range(len(right_servos_list)):
        right_servos_list[index]["index"] = index
        right_servos_list[index]["current"] = right_servos_list[index]["rest_pos"]
    right_servos_list = {right_servos_list[index]["name"]: right_servos_list[index] for index in range(len(right_servos_list))}
    
    return left_servos_list, right_servos_list

def get_servo_positions(servo_name, servo_list):
    min_pos = servo_list[servo_name]["min_pos"]
    max_pos = servo_list[servo_name]["max_pos"]
    rest_pos = servo_list[servo_name]["rest_pos"]
    servo_index = servo_list[servo_name]["index"]
    servo_current = servo_list[servo_name]["current"]
    
    return servo_index, rest_pos, min_pos, max_pos, servo_current
    

left_servos_file="../action/servos_data_left.csv"
right_servos_file="../action/servos_data_right.csv"

left_servos_list, right_servos_list = load_servo_info(left_servos_file, right_servos_file)

arm_servos = ("shoulder", "omoplate", "rotate", "bicep")

servo_info = dict()
servo_info["left"] = list()
servo_info["right"] = list()

for position in ("left", "right"):
    for servo_name in arm_servos:
        if position == "left":
            servos_list = left_servos_list
        if position == "right":
            servos_list = right_servos_list
            
        servo_index, _, min_pos, max_pos, _ = get_servo_positions(servo_name, servos_list)

        if not isinstance(min_pos, int):
            min_pos = int(min_pos)
        if not isinstance(max_pos, int):
            max_pos = int(max_pos)
        if not isinstance(servo_index, int):
            servo_index = int(servo_index)

        servo_info[position].append(
            {
                "servo_name":servo_name,
                "servo_index":servo_index,
                "min_pos":min_pos,
                "max_pos":max_pos
            }
        )       

# I will access the limits of the servos through this dictionary
servo_info 
```

```python
# Helper functions adapted from motion_controller.py

def initiate_connection(
    left_port="/dev/ttyACM0", 
    right_port="/dev/ttyACM1", 
    baud_rate=9600
):
    arduino_left = serial.Serial(left_port, baud_rate, timeout=1)
    arduino_right = serial.Serial(right_port, baud_rate, timeout=1)
    time.sleep(2)  # Give them time to initialize

    print("Connection has been successfully initiated!")
    return arduino_left, arduino_right

def close_connection(arduino_left, arduino_right):
    if arduino_left is not None and arduino_right is not None:
        time.sleep(5)
        arduino_left.close()
        arduino_right.close()
        print("Connection has been successfully closed!")
    else:
        print("Connection has not been initiated!")

def send_command(servo_name, target_angle, position, verbose=True):
    for servo_i in servo_info[position]:
        if servo_name == servo_i["servo_name"]:
            servo = servo_i
            break
        
    arduino = arduino_left if position == "left" else arduino_right

    target_angle = min(servo["max_pos"], max(target_angle, servo["min_pos"]))
    servo_index = servo["servo_index"]
    try:          
        arduino.write(f"{servo_index} {target_angle}\n".encode())
        if verbose:
            print(f"> Moving {servo_name} ({servo_index}) to {target_angle} on arduino {position}. <")    
            response = arduino.readline().decode().strip()
            print(f"- {position} arduino response:", response)
            
    except Exception as e:
        print(f"Error: {e}!")

# Utility to calculate angle between three points
def calculate_angle(a, b, c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)
    ba = a - b
    bc = c - b
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc) + 1e-6)
    angle = np.arccos(np.clip(cosine_angle, -1.0, 1.0))
    return np.degrees(angle)

# Utility to extract keypoints for shoulders, elbows, wrists
def extract_keypoints(landmarks, width, height):
    def get_point(lm):
        return [lm.x * width, lm.y * height]

    points = {}
    points['left_shoulder'] = get_point(landmarks[mp_holistic.PoseLandmark.LEFT_SHOULDER])
    points['left_elbow']    = get_point(landmarks[mp_holistic.PoseLandmark.LEFT_ELBOW])
    points['left_wrist']    = get_point(landmarks[mp_holistic.PoseLandmark.LEFT_WRIST])
    points['right_shoulder'] = get_point(landmarks[mp_holistic.PoseLandmark.RIGHT_SHOULDER])
    points['right_elbow']    = get_point(landmarks[mp_holistic.PoseLandmark.RIGHT_ELBOW])
    points['right_wrist']    = get_point(landmarks[mp_holistic.PoseLandmark.RIGHT_WRIST])
    return points

def normalize_and_scale_keypoints(kps: dict, img_w: int, img_h: int) -> dict:
    """
    Normalize keypoint coordinates to [0,1] range based on image size.
    Assumes each value in kps is a tuple/list (x, y, z).
    """
    norm_kps = {}
    for name, (x, y) in kps.items():
        norm_x = x / img_w
        norm_y = y / img_h
        norm_kps[name] = (norm_x, norm_y)
    return norm_kps

ARM_CHAINS = [
    ["left_shoulder", "left_elbow", "left_wrist"],
    ["right_shoulder", "right_elbow", "right_wrist"],
]

def plot_random_samples(data_file: str, n_samples: int = 3):
    # 1. Load
    with open(data_file, 'r') as f:
        data = json.load(f)

    # 2. Pick n_samples at random
    samples = random.sample(data, n_samples)

    # 3. Set up subplots
    fig, axes = plt.subplots(1, n_samples, figsize=(5 * n_samples, 5))

    for ax, sample in zip(axes, samples):        
        mp = sample['mp']           # normalized keypoints dict: name -> (x, y, z)
        angles = sample['angles']   # list of {"servo":…, "angle":…}
        sid = sample.get('id', '?')

        # Extract x,y coords and labels
        labels = list(mp.keys())
        xs = [coord[0] for coord in mp.values()]
        ys = [coord[1] for coord in mp.values()]

        # 4. Plot each keypoint and annotate
        for x, y, lbl in zip(xs, ys, labels):
            ax.scatter(x, y)
            ax.annotate(lbl, (x, y), fontsize=8, alpha=0.8)

        # 5. Build servo-angle text box
        servo_txt = '\n'.join(f"{d['servo']}: {d['angle']}" for d in angles)
        ax.text(
            0.02, 0.98,
            servo_txt,
            transform=ax.transAxes,
            fontsize=8,
            verticalalignment='top',
            bbox=dict(boxstyle='round,pad=0.3', alpha=0.3)
        )

        # 6. Joints
        for chain in ARM_CHAINS:
            points = [mp[k][:2] for k in chain if k in mp]
            if len(points) == len(chain):
                xs, ys = zip(*points)
                ax.plot(xs, ys, linewidth=1, color="black")

        ax.set_title(f"Sample {sid}")
        ax.set_xlim(0, 1)
        ax.set_ylim(1, 0)   # flip Y so “0,0” is top-left like an image
        ax.set_aspect('equal')

    plt.tight_layout()
    plt.show()

def list_cameras(max_tested=5):
    """
    Returns a list of indices for cameras that can be opened.
    By default, it checks camera indices 0 through 4.
    """
    available_cameras = []
    for index in range(max_tested):
        cap = cv2.VideoCapture(index)
        if cap.isOpened():
            available_cameras.append(index)
            cap.release()
    return available_cameras
    
def test_camera(index):
    """
    Opens the camera at `index`, displays a live feed,
    and waits until 'q' is pressed to quit.
    """
    cap = cv2.VideoCapture(index)
    if not cap.isOpened():
        print(f"Could not open camera {index}.")
        return

    print(f"Testing camera index {index}. Press 'q' to quit this camera.")
    while True:
        ret, frame = cap.read()
        if not ret:
            print(f"Failed to read frame from camera {index}.")
            break

        cv2.imshow(f"Camera {index}", frame)

        # Press 'q' to close the current camera window
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyWindow(f"Camera {index}")

def save_augmented_images(data_file, image_folder, output_folder):
    """
    For each sample in the JSON data, load the corresponding raw image from image_folder,
    overlay keypoints and bones, draw servo-angle box, and save to output_folder.
    
    Assumes images are named: sample_{id}.jpg (or .png).
    """
    os.makedirs(output_folder, exist_ok=True)
    with open(data_file, 'r') as f:
        data = json.load(f)

    for sample in data:
        sid = sample['id']
        angles = sample['angles']
        mp = sample['mp']
        
        # Load the original image
        img_path = os.path.join(image_folder, f"sample_{sid}.jpg")
        if not os.path.isfile(img_path):
            img_path = os.path.join(image_folder, f"sample_{sid}.png")
        img = Image.open(img_path)
        w, h = img.size
        
        fig, ax = plt.subplots(figsize=(6, 6))
        ax.imshow(img)
        
        # Overlay keypoints
        for name, (x, y) in mp.items():
            ax.scatter(x * w, y * h, s=20, color='cyan')
            ax.text(x * w, y * h, name, color='white', fontsize=6)
        
        # Draw bones
        for chain in ARM_CHAINS:
            pts = [(mp[k][0] * w, mp[k][1] * h) for k in chain if k in mp]
            if len(pts) == len(chain):
                xs, ys = zip(*pts)
                ax.plot(xs, ys, linewidth=2, color='lime')
        
        # Draw servo-angle box
        servo_txt = '\n'.join(f"{d['servo']}: {d['angle']}" for d in angles)
        props = dict(boxstyle='round,pad=0.3', facecolor='black', alpha=0.5)
        ax.text(0.02, 0.98, servo_txt, transform=ax.transAxes, fontsize=8, color='white',
                verticalalignment='top', bbox=props)
        
        ax.axis('off')
        save_path = os.path.join(output_folder, f"augmented_{sid}.png")
        plt.savefig(save_path, bbox_inches='tight')
        plt.close(fig)

def plot_random_samples_with_images(data_file: str,
                                    image_folder: str,
                                    n_samples: int = 3):
    """
    Pick n_samples at random from data_file, and for each one:
      • plot its pose (keypoints, chains, angles)
      • display the matching PNG from image_folder,
        assuming filenames like sample_<id>.png
    """
    # 1. Load pose data
    with open(data_file, 'r') as f:
        data = json.load(f)

    # 2. Randomly select
    samples = random.sample(data, n_samples)

    # 3. Prepare subplots: n_samples rows, 2 columns
    fig, axes = plt.subplots(n_samples, 2,
                             figsize=(10, 5 * n_samples))
    # if only one sample, make axes 2D
    if n_samples == 1:
        axes = axes.reshape(1, 2)

    for i, sample in enumerate(samples):
        sid = sample.get('id', '?')
        mp = sample['mp']
        angles = sample['angles']

        ax_pose, ax_img = axes[i]

        # --- Pose plot on the left ---
        # blank background
        ax_pose.set_xlim(0, 1)
        ax_pose.set_ylim(1, 0)
        ax_pose.set_aspect('equal')
        ax_pose.axis('off')

        # draw keypoints
        for lbl, (x, y, *_rest) in mp.items():
            ax_pose.scatter(x, y, s=20, c='red')
            ax_pose.annotate(lbl, (x, y),
                             fontsize=8, color='white',
                             ha='center', va='center', alpha=0.8)

        # draw joint chains
        for chain in ARM_CHAINS:
            pts = [(mp[k][0], mp[k][1]) for k in chain if k in mp]
            if len(pts) == len(chain):
                xs, ys = zip(*pts)
                ax_pose.plot(xs, ys, linewidth=2, color='yellow')

        # servo‐angle textbox
        servo_txt = '\n'.join(f"{d['servo']}: {d['angle']}"
                              for d in angles)
        ax_pose.text(0.02, 0.98, servo_txt,
                     transform=ax_pose.transAxes,
                     fontsize=8, color='white',
                     verticalalignment='top',
                     bbox=dict(boxstyle='round,pad=0.3',
                               facecolor='black', alpha=0.6))

        ax_pose.set_title(f"Pose {sid}", color='white')

        # --- Sample PNG on the right ---
        img_path = os.path.join(image_folder, f"augmented_{sid}.png")
        if os.path.exists(img_path):
            img = plt.imread(img_path)
            ax_img.imshow(img)
            ax_img.set_title(f"Image {sid}")
        else:
            ax_img.text(0.5, 0.5, f"Missing:\n{img_path}",
                        ha='center', va='center', color='red')
        ax_img.axis('off')

    plt.tight_layout()
    plt.show()
```

### 1. Data Collection

```python
# TODO

# 1. Generate random angles (withing each servo limits) for the servos on the shoulder and bicep.
# 2. Wait maybe 6 or 7 seconds to give the servos some time to position.
# 3. Use MediaPipe to capture the shoulder, elbow, and wrist positions.
# 4. Normalize and scale those MediaPipe positions. (Very important step because this should be the same process I should apply on my recordings later)
# 5. Save the random angles and the MediaPipe positions as one sample.
# 6. Repeat 1 - 5 an n amount of times.

# Hands to work ->

# 0. Initiate connections
NUM_SAMPLES = 1000
OUTPUT_FILE = "data_inverse_kinematics.json"
RAW_IMG_DIR = "raw_images"

# Prepare output directory
os.makedirs(RAW_IMG_DIR, exist_ok=True)

# Initialize connections
time.sleep(5)  # Time to step out of frame
arduino_left, arduino_right = initiate_connection()
cap = cv2.VideoCapture(4)

# MediaPipe setup
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic

samples = []
start_time = time.time()

with mp_holistic.Holistic(min_detection_confidence=0.5,
                          min_tracking_confidence=0.5) as holistic:
    try:
        for i in range(NUM_SAMPLES):
            print(f"\n-> Sample {i}")

            sample = {"id": i, "angles": [], "mp": None}

            # 1. Generate & send random angles within each servo's limits
            for side in ("left", "right"):
                for info in servo_info[side]:
                    name = info["servo_name"]
                    angle = np.random.randint(info["min_pos"], info["max_pos"] + 1)
                    sample["angles"].append({"servo": f"{name}_{side}", "angle": angle})
                    send_command(name, angle, side, verbose=True)
                    time.sleep(0.1)

            # 2. Allow servos to move and camera to see
            time.sleep(6)
            
            # 3. Capture frame
            if not cap.isOpened():
                raise RuntimeError("Camera not available")

            for _ in range(5):
                cap.read()
                
            cv2.waitKey(100) 
            ret, frame = cap.read()
            if not ret:
                print(f"Failed to grab frame for sample {i}")
                continue

            # Save raw image
            raw_path = os.path.join(RAW_IMG_DIR, f"sample_{i}.jpg")
            cv2.imwrite(raw_path, frame)

            # 4. Process with MediaPipe
            img_rgb = cv2.cvtColor(cv2.flip(frame, 1), cv2.COLOR_BGR2RGB)
            img_rgb.flags.writeable = False
            results = holistic.process(img_rgb)
            img_rgb.flags.writeable = True

            h, w, _ = frame.shape
            if results.pose_landmarks:
                raw_kps = extract_keypoints(results.pose_landmarks.landmark, w, h)
                sample["mp"] = normalize_and_scale_keypoints(raw_kps, w, h)
                # 5. Append sample
                samples.append(sample)
            else:
                print(f"No pose detected for sample {i}")

            
            

    finally:
        # 6. Cleanup and save
        cap.release()
        close_connection(arduino_left, arduino_right)

        with open(OUTPUT_FILE, "w") as f:
            json.dump(samples, f, indent=2)
        print(f"Saved {len(samples)} samples to {OUTPUT_FILE}")

end_time = time.time()
print(f"Total duration: {end_time - start_time:.2f} seconds")
save_augmented_images("data_inverse_kinematics.json", "raw_images", "augmented_images")
```

### 2. Model Definition, Training, and Testing

```python
# TODO

# based on the paper https://www.researchgate.net/publication/260014821_Neural_Network_based_Inverse_Kinematics_Solution_for_Trajectory_Tracking_of_a_Robotic_Arm
# Train a FNN (input:positions, 1 hidden layer 100 neurons, output:angles) using hyperbolic tangent. (These are the settings from the paper)
```

```python
# Imports
import json
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader, random_split
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
# Load the data and prepare it
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# 1. Load and prepare the data
with open("data/data_inverse_kinematics.json") as f:
    data = json.load(f)

# Define the feature and label order
joint_keys = ['left_shoulder', 'left_elbow', 'left_wrist',
              'right_shoulder', 'right_elbow', 'right_wrist']
servo_order = ['shoulder_left', 'omoplate_left', 'rotate_left', 'bicep_left',
               'shoulder_right', 'omoplate_right', 'rotate_right', 'bicep_right']

X, Y = [], []
for item in data:
    feats = []
    for k in joint_keys:
        feats.extend(item['mp'][k])
    X.append(feats)
    angle_dict = {a['servo']: a['angle'] for a in item['angles']}
    Y.append([angle_dict[s] for s in servo_order])

# Convert to torch tensors
X = torch.tensor(X, dtype=torch.float32)
Y = torch.tensor(Y, dtype=torch.float32)

# 2. Create Dataset and split (80/10/10)
class IKDataset(Dataset):
    def __init__(self, X, Y):
        self.X, self.Y = X, Y
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.Y[idx]

dataset = IKDataset(X, Y)
n = len(dataset)
train_n = int(0.8 * n)
val_n   = int(0.1 * n)
test_n  = n - train_n - val_n

train_ds, val_ds, test_ds = random_split(
    dataset, [train_n, val_n, test_n], generator=torch.Generator().manual_seed(42)
)

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_ds, batch_size=32)
test_loader  = DataLoader(test_ds, batch_size=32)
```

```python
X.shape, Y.shape
```

```python
# 3. Define the FNN model
class FNN(nn.Module):
    def __init__(self, input_dim=12, hidden_dim=100, output_dim=8):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, output_dim)
        )
    def forward(self, x):
        return self.net(x)
```

```python
# 4. Training and validation functions
def train_epoch(model, loader, criterion, optimizer):
    model.train()
    running_loss = 0.0
    for xb, yb in loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        preds = model(xb)
        loss = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * xb.size(0)
    return running_loss / len(loader.dataset)

def eval_epoch(model, loader, criterion, tol=5.0):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():
        for xb, yb in loader:
            xb, yb = xb.to(device), yb.to(device)
            preds = model(xb)
            loss = criterion(preds, yb)
            running_loss += loss.item() * xb.size(0)
            correct += (torch.abs(preds - yb) < tol).all(dim=1).sum().item()
            total += xb.size(0)
    return running_loss / len(loader.dataset), correct / total
```

```python
# 5. Run training
model = FNN().to(device)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)

num_epochs = 200
train_losses, val_losses, val_accs = [], [], []

for epoch in range(1, num_epochs+1):
    tl = train_epoch(model, train_loader, criterion, optimizer)
    vl, va = eval_epoch(model, val_loader, criterion)
    train_losses.append(tl)
    val_losses.append(vl)
    val_accs.append(va)
    print(f"Epoch {epoch:02d}/{num_epochs} – Train Loss: {tl:.4f}, Val Loss: {vl:.4f}, Val Acc: {va:.4f}")

# 6. Plot training vs. validation loss
plt.figure()
plt.plot(range(1, num_epochs+1), train_losses, label='Train Loss')
plt.plot(range(1, num_epochs+1), val_losses, label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.title('Training vs. Validation Loss')
plt.show()
```

```python
# 7. Save & reload the model
torch.save(model.state_dict(), 'ik_fnn.pth')

loaded_model = FNN().to(device)
loaded_model.load_state_dict(torch.load('ik_fnn.pth', map_location=device))
loaded_model.eval()
```

```python
# grab one random sample from the test set
idx = torch.randint(len(test_ds), (1,)).item()
x_sample, y_true = test_ds[idx]

# move to device and predict
x_sample = x_sample.to(device)
with torch.no_grad():
    y_pred = loaded_model(x_sample.unsqueeze(0)).squeeze(0).cpu()

print(f"Sample index: {idx}")
print("Actual angles:   ", y_true.tolist())
print("Predicted angles:", y_pred.tolist())

# 9. Plot actual vs. predicted
servo_names = servo_order  # must match the order you used for Y
x = np.arange(len(servo_names))

plt.figure()
plt.bar(x - 0.2, y_true.numpy(), width=0.4, label='Actual')
plt.bar(x + 0.2, y_pred.numpy(), width=0.4, label='Predicted')
plt.xticks(x, servo_names, rotation=45)
plt.ylabel('Angle (°)')
plt.legend()
plt.title('Actual vs. Predicted Servo Angles')
plt.tight_layout()
plt.show()
```
